```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return self.rank + " of "+self.suit
```

```python
class Deck:
    def __init__(self):
        self.allcards = []

        for suit in suits:
            for rank in ranks:
                created_card = Card(suit,rank)
                self.allcards.append(created_card)
    def shuffle(self):
        random.shuffle(self.allcards)
    
    def onecard(self):
        return self.allcards.pop() 
```

```python
class Player:
    def __init__(self,name):
        self.name = name

        self.allcards = []
        
    def add_one(self,new_cards):
        if type(new_cards) == type([]):
            # List of multiple card
            self.allcards.extend(new_cards)
        else:
            # List of Single card  
            self.allcards.append(new_cards)
    def remove_one(self):
        return self.allcards.pop(0)
    def __str__(self):
        return f'Player {self.name} has {len(self.allcards)} cards.'
```
