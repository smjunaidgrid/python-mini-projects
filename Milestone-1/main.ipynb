```python
def win_chk(board,mark):
    # Win check in rows or columns or diagonal checks 
    return(
    #Rows 
    (board[1] == board[2] == board[3] == mark) or (board[4]==board[5]==board[6] == mark) or (board[7] == board[8] == board[9]== mark)or
    #Columns
    (board[1] == board[4] == board[7] == mark) or (board[2]==board[5]==board[8] == mark) or (board[3] == board[6] == board[9]== mark)or
    #Diagonals
    (board[1] == board[5] == board[9] == mark) or (board[3]==board[5]==board[7] == mark)
    )  
```

```python
import random
def choose_player():
    flip = random.randint(0,1)
    if flip == 0:
        return "Player 1"
    else: 
        return "Player 2"
    
```

```python
def space_chk(board,position):
    return board[position]==' '
```

```python
def board_full(board):
    for i in range(1,10):
        if space_chk(board,i):
            return False
    return True
```

```python
def place_choice(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_chk(board,position):
        position = int(input("Choose integer from 1-9 : ")) 
    return position
```

```python
def replay():
    choice = input("Play Again? Enter Yes or No ")
    return choice == 'Yes'
```

```python
print("Welcome to Tic Tac Toe")
while True:
    the_board = [' ']*10
    player1_marker,player2_marker = player_input() 

    turn = choose_player()
    print(turn + "Will go first: ")

    play_game = input("Ready to play?  'y' or 'n' ")

    if play_game == 'yes':
        game_on = True
    else:
        game_on = False
    # GAME START
    while game_on:
        if turn == 'Player 1':
            display_board(the_board)
            #CHOICE 
            position = place_choice(the_board)
            place_marker(the_board, player1_marker, position)


        ##IF win 
            if win_chk(the_board,player1_marker):
                display_board(the_board)
                print("Congratulations Player 1 has won the game! ")
                game_on = False
            else:
                if board_full(the_board):
                    display_board(the_board)
                    print("Tie Game!")
                    game_on = False
                else:
                    turn = 'Player 2'
        else:
            if turn == 'Player 2':
                display_board(the_board)
                #CHOICE 
                position = place_choice(the_board,player2_marker,position)

            ##IF win 
            if win_chk(the_board,player2_marker):
                display_board(the_board)
                print("Congratulations Player 2 has won the game! ")
                game_on = False
            else:
                if board_full(the_board):
                    display_board(the_board)
                    print("Tie Game!")
                    game_on = False
                else:
                    turn = 'Player 1'
    if not replay():
        break
```
